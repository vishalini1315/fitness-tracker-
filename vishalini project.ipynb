```python
 age_calories_avg = df.groupby('Age')['Calories_Burned'].mean()
print(age_calories_avg)
```

```python
workout_freq_fat = df.groupby('Workout_Frequency (days/week)')['Fat_Percentage'].mean()
print(workout_freq_fat)
```

```python
workout_duration_calories = df.groupby('Workout_Type')[['Session_Duration (hours)', 'Calories_Burned']].mean()
print(workout_duration_calories)
```

```python
experience_resting_bpm = df.groupby('Experience_Level')['Resting_BPM'].mean()
print(experience_resting_bpm)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.hist(df['Calories_Burned'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Calories Burned')
plt.xlabel('Calories Burned')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df['Resting_BPM'], bins=20, color='salmon', edgecolor='black')
plt.title('Distribution of Resting Heart Rate (BPM)')
plt.xlabel('Resting BPM')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Session_Duration (hours)', y='Calories_Burned')
plt.title('Calories Burned vs Session Duration')
plt.xlabel('Session Duration (hours)')
plt.ylabel('Calories Burned')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Experience_Level', y='Resting_BPM')
plt.title('Resting BPM vs Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Resting BPM')
plt.show()
```

```python
selected_columns = ['Calories_Burned', 'Session_Duration (hours)', 'Avg_BPM', 'Resting_BPM', 'Fat_Percentage']
correlation_matrix = df[selected_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Key Fitness Metrics')
plt.show()
```
